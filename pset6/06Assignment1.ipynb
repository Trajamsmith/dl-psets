```python
import cv2
import matplotlib.pyplot as plt

def cv2ToPlt(cv2Img):
    return plt.imshow(cv2.cvtColor(cv2Img, cv2.COLOR_BGR2RGB))
```

```python
camel = cv2.imread('camel.jpg', 1)
cv2ToPlt(camel)
```

```python
dunes = cv2.imread('dunes.jpg', 1)
cv2ToPlt(dunes)
```

```python
blend = cv2.add(camel, dunes, 0)
cv2ToPlt(blend)
```

With raw image addition, things get washed out, which make sense.

```python
blend5050 = cv2.addWeighted(camel, 0.5, dunes, 0.5, 0)
cv2ToPlt(blend5050)
```

A 50-50 weighted blend looks better. We're no longer washing out the pixels through basic integer additions. Now let's try some different blend levels below:

```python
blend9010 = cv2.addWeighted(camel, 0.9, dunes, 0.1, 0)
cv2ToPlt(blend9010)
```

```python
blend = cv2.addWeighted(camel, 0.6, dunes, 0.4, 0)
cv2ToPlt(blend)
```

```python
blend1090 = cv2.addWeighted(camel, 0.1, dunes, 0.9, 0)
cv2ToPlt(blend1090)
```

```python
blend = cv2.addWeighted(camel, 0.4, dunes, 0.6, 0)
cv2ToPlt(blend)
```

```python
subtracted = cv2.subtract(blend5050, camel)
cv2ToPlt(subtracted)
```

Subtracting images gives us these crazy color inversions. Let's try it on another blend for fun.

```python
subtracted = cv2.subtract(blend1090, camel)
cv2ToPlt(subtracted)
```

```python
subtracted = cv2.subtract(blend9010, camel)
cv2ToPlt(subtracted)
```

Subtracting the camel image from a blend that is almost entirely camel results in a mostly empty image. This makes sense intuitivel.
