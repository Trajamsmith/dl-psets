## Problem 2

Borrowing code from the example notebook:

```python
import numpy as np
import gensim

# Get the interactive Tools for Matplotlib
%matplotlib notebook
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')

from sklearn.decomposition import PCA
from gensim.test.utils import datapath, get_tmpfile
from gensim.models import KeyedVectors
from gensim.scripts.glove2word2vec import glove2word2vec
```

```python
import os
cwd = os.getcwd()

glove_file = datapath(cwd + '\glove.6B\glove.6B.100d.txt')
word2vec_glove_file = get_tmpfile("glove.6B.100d.word2vec.txt")
glove2word2vec(glove_file, word2vec_glove_file)
```

```python
model = KeyedVectors.load_word2vec_format(word2vec_glove_file)
```

```python
def display_pca_scatterplot(model, words=None, sample=0):
    if words == None:
        if sample > 0:
            words = np.random.choice(list(model.vocab.keys()), sample)
        else:
            words = [ word for word in model.vocab ]
            
    word_vectors = np.array([model[w] for w in words])
    twodim = PCA().fit_transform(word_vectors)[:,:2]
    
    plt.figure(figsize=(6,6))
    plt.scatter(twodim[:,0], twodim[:,1], edgecolors='k', c='r')
    
    for word, (x,y) in zip(words, twodim):
        plt.text(x+0.05, y+0.05, word)
        
    for i in range(0, len(twodim), 2):
        plt.plot(twodim[:,0][i:i+2], twodim[:,1][i:i+2], 'bo-')
```

```python
display_pca_scatterplot(model, ['france','paris','russia','moscow','greece','athens','england','london'])
```

We can see that the vectors are roughly parallel, showing similar relationships between the words (as expected).

```python
new_analogies = ['chef', 'restaurant', 
                 'doctor', 'hospital', 
                 'teacher', 'school', 
                 'professor', 'university', 
                 'trainer', 'gym',
                 'manager', 'office',
                 'artist', 'studio',
                 'actor', 'theater',
                ]

display_pca_scatterplot(model, new_analogies)
```

This fun example looks pretty good. The vectors are very roughly parallel across all eight pairs. Artist-Studio and Manager-Office being the most off-axis makes sense to me (I think they're the most tenuous analogies).

## Problem 4
Now let's try using t-SNE to visualize our data, rather than PCA:

```python
from sklearn.manifold import TSNE

def display_tsne_scatterplot(model, words=None, sample=0):
    if words == None:
        if sample > 0:
            words = np.random.choice(list(model.vocab.keys()), sample)
        else:
            words = [ word for word in model.vocab ]
            
    word_vectors = np.array([model[w] for w in words])
    twodim = TSNE().fit_transform(word_vectors)[:,:2]
    
    plt.figure(figsize=(6,6))
    plt.scatter(twodim[:,0], twodim[:,1], edgecolors='k', c='r')
    
    for word, (x,y) in zip(words, twodim):
        plt.text(x+0.05, y+0.05, word)
        
    for i in range(0, len(twodim), 2):
        plt.plot(twodim[:,0][i:i+2], twodim[:,1][i:i+2], 'bo-')
```

```python
display_tsne_scatterplot(model_300, ['france','paris','russia','moscow','greece','athens','england','london'])
```

This also looks very different but we're using an entirely different dimensionality reduction algorithm. This might be in line with what we should expect with t-SNE. Note that again, every vector except England-London is parallel.

## Problem 5
Let's try running our analogy experiments using spaCy:

```python
import spacy

nlp = spacy.load('en_core_web_lg')  # make sure to use larger model!
tokens = nlp(u'france paris russia moscow greece athens england london')
```

```python
# Saving tokens as variables
france = tokens[0]
paris = tokens[1]
russia = tokens[2]
moscow = tokens[3]
greece = tokens[4]
athens = tokens[5]
england = tokens[6]
london = tokens[7]

print(france.similarity(paris))
print(russia.similarity(moscow))
print(greece.similarity(athens))
print(england.similarity(london))
```

These look pretty good. Let's try a baseline to check

```python
print(france.similarity(moscow))
print(greece.similarity(london))
print(london.similarity(russia))
```

These values are worse, as expected!

```python
# We need to modify our scatterplot to work with spaCy
def spacy_pca_scatterplot(tokens, sample=0):
    words = [token.text for token in tokens]
    vectors = [token.vector for token in tokens]
    
    twodim = PCA().fit_transform(vectors)[:,:2]
    
    plt.figure(figsize=(6,6))
    plt.scatter(twodim[:,0], twodim[:,1], edgecolors='k', c='r')
    
    for word, (x,y) in zip(words, twodim):
        plt.text(x+0.05, y+0.05, word)
        
    for i in range(0, len(twodim), 2):
        plt.plot(twodim[:,0][i:i+2], twodim[:,1][i:i+2], 'bo-')
```

```python
spacy_pca_scatterplot(tokens)
```

This looks a lot like our results for our 300-dimensional gensim model, which is a good sign! Note that three of the pairs are again parallel, with just England-London perpendicular. 

If we look all the way back to Problem 1, England-London seemed to have the lowest similarity using the Google Word2Vec script as well. I wonder why. Nevertheless, all the results seem coherent after accounting for that small oddity.
